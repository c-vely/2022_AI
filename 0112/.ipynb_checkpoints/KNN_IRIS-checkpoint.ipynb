```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()

print("iris_dataset의 Keys : \n{}\n".format(iris_dataset.keys()))
print("key 예시(target_names) : \n{}\n".format(iris_dataset['target_names']))
print("key 예시(featrue_names) : \n{}\n".format(iris_dataset['feature_names']))

print("총 Data의 수 : {}\n".format(iris_dataset.data.shape))
print("첫번째 데이터의 예시 : 이름-{}, 범주-{}, 특징-{}\n".format(iris_dataset.target_names[0], iris_dataset.target[0], iris_dataset.data[0]))
```

```python
import random

num_iris_datas, _ = iris_dataset.data.shape
num_samples = 15
rand_sample_lndex = list(map(lambda x: random.randint(x*10, (x+1)*10-1), range(num_samples)))
test_sample = [[iris_dataset.target[x], iris_dataset.data[x]] for x in rand_sample_lndex]

org_sample = []
for i in range(num_iris_datas):
     if i not in rand_sample_lndex:
         org_sample.append([iris_dataset.target[i], iris_dataset.data[i]])

print("샘플 Index : {}\n".format(rand_sample_lndex))
print("Test 샘플수: {}, 남은 데이터수: {}\n".format(len(test_sample), len(org_sample)))
print("데이터 포맷 [Index, 붓꽃 특징] : {}".format(test_sample[0]))
```

```python
import numpy as np

def distance(p1,p2):
    dist = list(map(lambda x, y: (x-y)*(x-y), p1, p2))
    return np.sqrt(np.sum(dist))

calc_dist=[]
for t_s in test_sample:
    t_s_dist = []
    for o_s in org_sample:
        dist = distance(t_s[1], o_s[1])
        t_s_dist.append([o_s[0], dist])
    calc_dist.append(t_s_dist)

print("거리 결과 예 [index, 거리차이] : {}\n".format(calc_dist[0][0]))
```

```python
import matplotlib.pyplot as plt

def get_x_y_data(datas):
    x = list(map(lambda x: x[0], datas))
    y = list(map(lambda x: x[1], datas))
    return x, y

def show_dist(dist_data, test_sample):
    fig = plt.figure
    fig, axes = plt.subplots(5,3, figsize=(15, 20))
    index = 0
    for row in axes:
        for col in row:
            x, y = get_x_y_data(dist_data[index])
            sub_plot = col.scatter(x, y, marker='*', c=x)
            col.set_xticks([0,1,2])
            col.set_title("Iris index(GT): {}".format(test_sample[index][0]))
            index += 1            
        cb = fig.colorbar(sub_plot, ax=col, label='species')
        cb.set_ticks([0,1,2])
        cb.set_ticklabels(["setosa","versicolor","virginica"])
    plt.show()

show_dist(calc_dist, test_sample)
```

```python
for t_s in range(len(calc_dist)):
    len_t_s = len(calc_dist[t_s])
    for i in range(len_t_s):
        for j in range(i + 1, len_t_s):
            if calc_dist[t_s][j][1] < calc_dist[t_s][i][1]:
                temp = calc_dist[t_s][j]
                calc_dist[t_s][j] = calc_dist[t_s][i]
                calc_dist[t_s][i] = temp

def show_sorted_dist(dist_data, test_sample):
    fig = plt.figure
    fig, axes = plt.subplots(5,3, figsize=(20, 20))
    index = 0
    for row in axes:
        for col in row:
            x, y = get_x_y_data(dist_data[index])            
            sub_plot = col.scatter(np.arange(len(x)), y, s = 4, marker='*', c=x)
            sub_plot = col.scatter(np.arange(len(x)), x, marker='^', c=x)
            col.set_title("Iris index(GT): {}".format(test_sample[index][0]))
            index += 1            
        cb = fig.colorbar(sub_plot, ax=col, label='species')
        cb.set_ticks(np.arange(len(x)))
        cb.set_ticklabels(["setosa","versicolor","virginica"])
    plt.show()

show_sorted_dist(calc_dist, test_sample)
```

```python
res_knn = []
range_K = 100

for t_s in range(len(calc_dist)):
    res = []
    for i in range(range_K):
        count_index = [0, 0, 0]
        for j in range(i+1):
            if calc_dist[t_s][j][0] == 0:
                count_index[0] += 1
            elif calc_dist[t_s][j][0] == 1:
                count_index[1] += 1
            else:
                count_index[2] += 1
        label = 0
        if count_index[1] >= count_index[0] or count_index[2] >= count_index[0]:
            if count_index[1] >= count_index[2]:
                label = 1
            else:
                label = 2
        res.append(label)
    res_knn.append(res)

def check_same(v1, v2):
    if v1 == v2:
        return 1
    else:
        return 0

def check_res(pred, gt):
    return list(map(lambda x, y: check_same(x, y), pred, gt))

def show_kNN_res(knn_data, test_sample):
    test_gt = [x[0] for x in test_sample]
    fig = plt.figure
    fig, axes = plt.subplots(5,3, figsize=(20, 20))
    index = 0
    for row in axes:
        for col in row:
            cc = check_res(np.ones(range_K)*test_gt[index], knn_data[index])
            sub_plot = col.scatter(np.arange(len(cc)), cc, s = 5, marker='*', c = cc)
            col.set_yticks([0,1])
            col.set_title("Iris index(GT): {}".format(test_sample[index][0]))            
            cb = fig.colorbar(sub_plot, ax=col, label='KNN Result')
            cb.set_ticks([0, 1])
            cb.set_ticklabels(["Wrong", "Correct"])
            index += 1
    plt.show()

show_kNN_res(res_knn, test_sample)
```
